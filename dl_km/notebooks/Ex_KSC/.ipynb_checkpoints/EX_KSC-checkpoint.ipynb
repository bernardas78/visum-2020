```python
# Load the necessary packages
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
from skimage import io
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from scipy.ndimage import generic_filter
import cv2
from sklearn.metrics.pairwise import chi2_kernel
from sklearn.neighbors import NearestNeighbors
from scipy.spatial.distance import cdist
```

```python
# Load data
X = io.imread('church.png') / 255
plt.imshow(X)

# Parameters
scale_percent = 50 # percent of original size

# Resize image
width = int(X.shape[1] * scale_percent / 100)
height = int(X.shape[0] * scale_percent / 100)
dim = (width, height)
X = cv2.resize(X, dim)
```

```python
# Helper functions to create kernel

def rgb2ind(X,n_colors):
    "Color Quantization using K-Means"
    w, h, d = original_shape = tuple(X.shape)
    image_array = np.reshape(X, (w * h, d))
    #Fitting model on a small sub-sample of the data")
    image_array_sample = shuffle(image_array, random_state=0)[:1000]
    kmeans = KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample)
    labels = kmeans.predict(image_array)
    codebook = kmeans.cluster_centers_
    # Random codebook
    # codebook = shuffle(image_array, random_state=0)[:n_colors]
    # labels = pairwise_distances_argmin(codebook_random,image_array,axis=0)
    return codebook, labels
    
def recreate_image(codebook, labels, w, h):
    "Recreate the (compressed) image from the code book & labels"
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image

def ImgLocalHist(X,codebook):
    X = X.flatten()
    N = np.shape(X)[0]
    nmbColor = np.shape(codebook)[0]
    bins = np.arange(nmbColor+1)
    hist_val = np.histogram(X, bins = bins)
    return hist_val[0] / N

def movingWindow(image,codebook):
    w, h =  tuple(image.shape)
    N = w * h
    nmbColor = np.shape(codebook)[0]
    result = np.zeros((N,nmbColor))
    counter = 0
    for x in range(w):
        for y in range(h):
            idx_Xmin = max(0,x-1)
            idx_Xmax = min(w,x+1)
            idx_Ymin = max(0,y-1)
            idx_Ymax = min(h,y+1)
            window = image[idx_Xmin:idx_Xmax+1,idx_Ymin:idx_Ymax+1]
            hist_val = ImgLocalHist(window,codebook)
            result[counter,:] = hist_val
            counter += 1   
    return result
    

def img2localHistogram(X,codebook):
    "Create color histogram"
    localHist = movingWindow(X,codebook)
    return localHist

```

```python
# The image is quantized in the desired number of colors using K-means. 
# This results in 'n_colors' feature vector for each pixel.

# Parameters
n_colors = 8 # quantization levels

# Quantize the image
w, h, d = original_shape = tuple(X.shape)
codebook, labels = rgb2ind(X,n_colors)
X_quant = recreate_image(codebook, labels, w, h)

fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
plt.imshow(X)
a.set_title('Original Image')
b = fig.add_subplot(1, 2, 2)
plt.imshow(X_quant)
b.set_title('Quantized Image')

# Create a local color histogram by a moving window of size 3 x 3.
# This moving window smooths the colors over the image.
localHist = img2localHistogram(np.reshape(labels,(w,h)),codebook)
X_orig = X
```

```python
# The local color histogram is used to make a chi-squared kernel with bandwidth parameter 'gamma'. 
# The kernel spectral clustering routine makes use of this kernel to segment the image in 'n_clusters' groups.
from sklearn.cluster import SpectralClustering

# Parameters
gamma = 20 # Value for Chi squared kernel 
n_clusters = 5 # Number of clusters

# Run KSC
X = localHist
clustering = SpectralClustering(n_clusters=n_clusters, assign_labels='discretize',affinity='chi2',gamma = gamma).fit(X)
label_im = np.reshape(clustering.labels_,(w,h))

# Visualize Image segmentation
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
plt.imshow(X_orig)
a.set_title('Original Image')
b = fig.add_subplot(1, 2, 2)
plt.imshow(label_im, cmap='tab10')
b.set_title('Image Segmentation')

```
